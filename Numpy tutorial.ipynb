```python
# Notes
import numpy as np

# Make a 2x2 Numpy Array
create_zeros = np.zeros((2, 2))
print(create_zeros)

# Numpy Range Function
rangE = np.arange(5, 12) # Print from 5 through 11
print(rangE)

# Create random numbers ranging through numbers
rangE = np.random.randint(low=1, high=200, size=5)
print(rangE)

# Create random floats from 0.0 to 1.0
rangE = np.random.random(5)
print(rangE)
```

```python
# Create a Linear Dataset
import numpy

feature = np.arange(1, 21) # Create an array ranging from 1 to 20
print(f"Feature: {feature}")

label = (feature*3) + 4 # An example linear would be y = (x*3) + 4
print(f"Label: {label}")

# Create Noise in the dataset, ranging from -2 to 2, 20 in exact for the labels
noise = (np.random.random([20]) * 4) - 2
print(f"Noise: {noise}")

# Add noise to the label
label = label + noise
print(f"Noised Label: {label}")
```
